```python
from gpiozero import LED
from time import sleep
import cv2
import numpy as np
import random
import os

#global count
#count=0


# Cam Function
def camera(lane):
    global count
    count=0
    global img_counter
    
    if lane==0:                                 #temp code..remove later
        cam = cv2.VideoCapture(lane)           #choose camera
    else:
        cam = cv2.VideoCapture("/home/pi/Desktop/project/Haar Cascade Algorithm/video1.mp4")
    
    

    while True:
        ret, image = cam.read()
        if (ret == False):
            print("failed to grab frame")
            break

        #cv2.imshow('Imagetest',image)                   #remove this line later
        folder="fold"+str(lane)
        imgpth=os.path.join(folder, str(img_counter) + '.jpg')
        cv2.imwrite(imgpth, image)             
        
        img_name = "frame.png".format(img_counter)
        img_counter+=1
        print(img_name)
        
        #print("screenshot")
        break
        
    image = cv2.imread(imgpth)
    
    grey = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    car_cascade = cv2.CascadeClassifier('cars.xml')
    cars = car_cascade.detectMultiScale(grey, 1.1, 1)
    
    for (x, y, w, h) in cars:
        
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)
        count += 1
        
    print("Vehicle Count: ", count)
    #cv2.imwrite(imgpth,image)         #hide later maybe
 
    cv2.imshow("Lane"+str(lane), image)                 #remove this line later

    cam.release()
    cv2.destroyAllWindows()
    return count



def timecalc(count):
    i=count
    low=10
    medium=20
    high=30
    #super_high=40
    delay=0
    print("Number of vehicles= "+ str(i))
    if i < low:
        if i<1:
            print('green light delay = 0 sec')
        elif i<4:
            delay=i*1.5
            print('green light delay = '+str(delay)+' sec')
        elif i<7:
            delay=5*1.5
            print('green light delay = '+str(delay)+' sec')
        else:
            delay=7*1.5
            print('green light delay = '+str(delay)+' sec')
    elif i < medium:
        if i<medium/2:
            delay=i*1.5
            print('green light delay = '+str(delay)+' sec')
        else:
            delay=i*1.5
            print('green light delay = '+str(delay)+' sec')
    elif i < high:
        if i<high/2:
            delay=i*1.5
            print('green light delay = '+str(delay)+' sec')
        else:
            delay=60
            print('green light delay = '+str(delay)+' sec')
    return(delay)











# Lane Functions
def Lane0(count):
    
    duration= timecalc(count)
    
  
    red0.off()
    red1.off()
    red2.off()
    sleep(0.5)
    
    #red3.on()
    #amber0.on()
    #sleep(3)
    
    red0.off()
    amber3.off()
    red3.on()
    red1.on()
    red2.on()
    green0.on()
    sleep(duration)

    green0.off()
    amber0.on()
    sleep(3)
    
    count=camera(1)
    
    amber0.off()
    red0.on()
    
    
    
    #sleep(5)
    
    #amber2.on(5)
    #amber3.on(5)
    #sleep(3)
    #green kathikanam of lane 0 for "duration" seconds
    #yellow kathikanam of lane 0 for 5 sec
    
    
    
    #red 0 kathikanam 
   
    
   
   
    
    #cv2.waitKey(8000)
    

def Lane1(count):


    duration= timecalc(count)
    
    
    red1.off()
    red2.off()
    red3.off()
    sleep(0.5)
    
    #red0.on()
    #amber1.on()
    #sleep(3)
    
    red1.off()
    
    amber0.off()
    red0.on()
    green1.on()
    red2.on()
    red3.on()
    sleep(duration)

    green1.off()
    amber1.on()
    sleep(3)
    count=camera(2)
    
    
    
    #green kathikanam of lane 1
    #yellow kathikanam of lane 1
    
    
   
    

def Lane2(count):
    
    duration= timecalc(count)
    
    red2.off()
    red0.off()
    red3.off()
    sleep(0.5)
    
    #red1.on()
    #amber2.on()
    #sleep(3)
    
    red2.off()
    amber1.off()
    red1.on()
    green2.on()
    red0.on()
    red3.on()
    sleep(duration)

    green2.off()
    amber2.on()
    sleep(3)
    count=camera(3)
    
    
    
    
    #green kathikanam of lane 2
    #yellow kathikanam of lane 2
  

def Lane3(count):
    duration= timecalc(count)
    
    
    red3.off()
    red0.off()
    red1.off()
    sleep(0.5)
    
    #red2.on()
    #amber3.on()
    #sleep(3)
    
    amber2.off()
    red3.off()
    red2.on()
    green3.on()
    red0.on()
    red1.on()
    sleep(duration)

    green3.off()
    amber3.on()
    sleep(3)
    
    count=camera(0)
    
    #sleep(3)
    #green kathikanam of lane 3
    #yellow kathikanam of lane 3
    





    
red0 = LED(16)
red1 = LED(26)                          
red2 = LED(10)               #change from 4 to 27 (prev err) to 10
red3 = LED(6)               #change from 25 to 6

amber0 = LED(19)
amber1 = LED(20)            #amber 1 and 2 pin interchanged
amber2 = LED(12)             #change to 12 from 3
amber3 = LED(5)              #change from 8 to 5

green0 = LED(13)
green1 = LED(21)            #green 1 and 2 pin interchanged
green2 = LED(9)             #change from 2 to 24 to 22 (prev err) to 9
green3 = LED(17)             #change from 7 to 17


img_counter=0
camera(0)
while True:
    
    Lane0(count)
    Lane1(count)
    Lane2(count)
    Lane3(count)
    
```
